## handling missing values
- 'df.isnull()' - to detect missing values (same as df.isna())
- 'df.fillna()' - to fill missing values
-df.dropna() -to remove missing values
- df.replace - to replace or impute the values(Imputation)

```python
import pandas as pd
import numpy as np
```

```python
employee_list = [['James', 36, 75, 5428000],
               ['Villers', 38, 74, 3428000],
               ['VKole', 31, 70, 8428000],
               ['Smith', 34, 80, 4428000],
               ['Gayle', 40, 100, 4528000],
               ['Adam', 40, np.nan, 4528000],
               ['Rooter', 33, 72, 7028000],
               ['Peterson', 42, 85, 2528000],
               ['lynda', 42, 85, np.nan],
               [np.nan, 42, 85, np.nan],
               ['Jenny', np.nan, 100, 25632],
               ['Kenn', np.nan, 110, 25632],
                ['Aly', np.nan, 90, 25582],
               ['John', 41, 85, 1528000]]

emp_df = pd.DataFrame(employee_list,columns =['Name','Age','weight','Salary'])
emp_df.head(10)
```

```python
rows,cols =emp_df.shape
print('rows',rows ,'Column',cols)
```

```python
emp_df.isnull().sum()
```

# Be wary of missing values,if u hav small dataset ,removing values can alter analysis
# missing data rows is less than 2.5 % of data set is safe.
# when dropping na, you remove the entire row

```python
dropped_df = emp_df.dropna()
print(dropped_df)
```

```python
filled_df =emp_df.fillna('None')
print(filled_df)
```

```python
filled_df.isnull().sum()
```

```python
# only change values of certain columns
specific_cols = emp_df[['weight','Age']].fillna('pending')
specific_cols
```

```python
emp_df2 = emp_df.fillna(value ={'weight':'pending','Age':'unknown','Salary':'0.0','Name':'verification pending'})
emp_df2
```

```python
technologies = {
    'Courses':["Spark",np.nan,"Hadoop","Python","pandas",np.nan,"Java"],
    'Fee' :[20000,25000, np.nan,22000,24000,np.nan,22000],
    'Duration':[np.nan,'40days','35days', np.nan,'60days','50days','55days'],
    'Discount':[1000,np.nan,1500,np.nan,2500,2100,np.nan]
              }

tech_df = pd.DataFrame(technologies)
tech_df.head(10)
```

```python
rows,cols= tech_df.shape
print('rows',rows ,'Column',cols)
```

```python
tech_df.isnull().sum()
```

```python
tech1_df =tech_df.replace(np.nan,'')
tech1_df
```

```python
tech3= tech_df.replace({'Courses':np.nan,'Fee':np.nan},'Pending')
tech3
```

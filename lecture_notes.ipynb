## pandas datastructure
- series - single dim,similar to list or single column
homogenous
- data frame- multi dim heterogenous

```python
ser1 = pd.Series([10,20,30,40] ,index=["A","B","C","D"])
print(ser1)
```

```python
# creating a series from a preexisting dictionary
d = {"b":1,
     "a":0,
     "c":2,
     "d":100}
ser2 = pd.Series(d)
print(ser2)
```

```python
d1 = {"a":0.0,
      "b":1.0,
      "c":2.0}
ser3 = pd.Series(d1)
print(ser3)
```

```python
calories ={
    'day1': 700,
    'day2':2500,
    'day3':1200
}
ser4 = pd.Series(calories,index=['day1','day3'])
print(ser4)
```

```python
ser5 =pd.Series(5.0, index=["a","b","c"])
print(ser5)
```

## Frequency Analysis - Value counts

```python
# Sample DataFrame
data = {'Category': ['Electronics', 'Clothing', 'Electronics', 'Books', 'Books', 'Clothing']}
cat_df = pd.DataFrame(data)
cat_df.value_counts()
```

```python
# Create the dataframe
sales_data={"Devices":['Laptop','iPhone','LED','LCD','Smart-Phone','Washing-Machine'],
           'Brand':['Lenovo','Apple','Samsung','Samsung','Samsung','Whirpool'],
           'Sales':[1000,2000,4000,2000,1000,4000],
           'Profit':[500,1000,1000,1500,1000,1500],
           'Pices left':[5000,4000,4000,5000,5000,1000]}
sales_df = pd.DataFrame(sales_data)
frequency = sales_df['Brand'].value_counts()
print(frequency)
freq_Product=frequency[frequency>1].index[0] 
print(freq_Product)
print('This item appear more than once')
```

```python
data = {'score':[85,92,88,75,None ,98,None,85]}
score_df = pd.DataFrame(data)
```

```python
score_counts = score_df.value_counts(dropna=False)
print(score_counts)
```

## pandas datastructure
- series - single dim,similar to list or single column
homogenous
- data frame- multi dim heterogenous

```python
ser1 = pd.Series([10,20,30,40] ,index=["A","B","C","D"])
print(ser1)
```

```python
# creating a series from a preexisting dictionary
d = {"b":1,
     "a":0,
     "c":2,
     "d":100}
ser2 = pd.Series(d)
print(ser2)
```

```python
d1 = {"a":0.0,
      "b":1.0,
      "c":2.0}
ser3 = pd.Series(d1)
print(ser3)
```

```python
calories ={
    'day1': 700,
    'day2':2500,
    'day3':1200
}
ser4 = pd.Series(calories,index=['day1','day3'])
print(ser4)
```

```python
ser5 =pd.Series(5.0, index=["a","b","c"])
print(ser5)
```

```python
## Data frame:
- indexes (each record) 1,2,3
- columns - (each variable) A B C
```

```python
fruit_data = {
    'apple_sales': [3, 2, 0, 2, 9, 4, 7, 2],
    'orange_sales': [0, 3, 2, 7, 7, 8, 3, 0 ]
}
idx =pd.RangeIndex(2000,2008)
purchases = pd.DataFrame(fruit_data,index =idx)
print(purchases)
```

```python
rows,cols = purchases.shape
print('rows',rows)
print('No of cols',cols)
```

```python
print(purchases.head())
```

```python
print(purchases.tail())
```

Adding columns to a DataFrame

```python
import numpy as np
import pandas as pd
```

```python
data = {'Name': ['Jane', 'Princi', 'James', 'Fadi'],
        'Height': [5.1, 6.2, 5.1, 5.2],
        'Qualification': ['Msc', 'MA', 'Msc', 'Msc'],
        'Score 1' : [56,86,77,45],
        'Score 2' : [50,96,60,30]}

df = pd.DataFrame(data)
```

```python
print('DataFrame before New column')
df.head()
```

```python
address = ['NYC','NJ','CA','PA']
df['Address'] = address
```

```python
print('Data frame after')
df.head()
```

Creatind a New column using an old column(s)

```python
df['Total_Score'] = df['Score 1'] +df['Score 2']
print("after column added")
df.head()
```

```python
df['avg_Score'] = (df['Total_Score'])/2
print("after column added")
df.head()
```

Importing specified columns and rearranging them

```python
df2 = pd.DataFrame(data,columns= ['Score 1','Score 2','Name','Qualification'])

print(df2)
```

```python
values = [['Rohan', 455], ['Elvish', 250], ['John', 495],
          ['Sai', 400], ['Eric', 350], ['Adam', 450]]
df3 = pd.DataFrame(values,columns=['Name','Univ_Marks'])
df3.head()
```

```python
df3['Percentage'] =(df3['Univ_Marks'] /500) * 100
df3.head()
```
